```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import diags
```

```python
b=np.array([1,1])
x=np.linspace(-10,10,1000)
y=np.linspace(-10,10,1000)
X,Y=np.meshgrid(x,y)
h=1/3
J=(1/(h**2))*(X**2+Y**2-X*Y)-(X+Y)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
#plot the surface
surf=ax.plot_surface(X,Y,J)
```
